# Good Reads Books Dataframe

```python
from bs4 import BeautifulSoup as bs
import urllib.request
import requests
import re
import time
import pandas as pd
```

```python
url_2018 = 'https://www.goodreads.com/list/show/129936.NYT_100_Notable_Books_of_2018'
r = urllib.request.urlopen(url_2018)

soup = bs(r, 'html.parser')
```

```python
soup
```

```python
coolcats = soup.find_all('div', {"class":"js-tooltipTrigger tooltipTrigger"})[0].a['href']
```
