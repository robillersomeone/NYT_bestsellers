```python
# method to get request all the genres
# this is where we change the genre
def make_urls_by_genre():
    genre_urls = []
    for i in monthly_genres:
        link = 'https://api.nytimes.com/svc/books/v3/lists/'+i+'/2019-01-06.json?api-key=f945e21becaa4beabcd4c92155d99506'
        genre_urls.append(link)
    return genre_urls
```

```python
make_urls_by_genre()
```

```python
# will break up into genres
def api_calls_by_genre():
    mini_list = []
    for i in make_urls_by_genre():
        r = requests.get(i)
        mini_list.append(r.json())
    return mini_list
```
